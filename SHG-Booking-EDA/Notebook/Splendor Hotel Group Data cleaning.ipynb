# 1)since my geographical analysis heavily depends on it, i filled my missing values with "Unknown"

```python
SHG['Country'].fillna('Unknown', inplace=True)
```

# 2) Converted my Date Column to Data Time because they are currently stored as object
# (i.e Just plain text or string format) which limit what i can do with them


```python
SHG['Booking Date'] = pd.to_datetime(SHG['Booking Date'])
SHG['Arrival Date'] = pd.to_datetime(SHG['Arrival Date'])
SHG['Status Update'] = pd.to_datetime(SHG['Status Update'])
```

```python
SHG
```

# 3) I Checked for Duplicates
# To Ensure there are no repeated records.

```python
SHG.duplicated().sum()
SHG.drop_duplicates(inplace=True)
```

# 4) Validating and Fixing my Data Types

## a) converted my Booking ID → string
### Why: Because i'm not performing math on it, Treating it as a string prevents accidental numerical operations.

```python
SHG['Booking ID'] = SHG['Booking ID'].astype(str)
```

## b) Converted Avg Daily Rate, Revenue, Revenue Loss → float
### Why: These are monetary values. Keeping them as int64 might hide decimals For price-related analysis, float ensures precision

```python
SHG['Avg Daily Rate'] = SHG['Avg Daily Rate'].astype(float)
SHG['Revenue'] = SHG['Revenue'].astype(float)
SHG['Revenue Loss'] = SHG['Revenue Loss'].astype(float)
```

## c) Lead Time, Nights, Guests → int is OK
### Why: These are count-based metrics. Integer is appropriate.

#### I Checked: Ensure no negative values:

```python
SHG[['Lead Time', 'Nights', 'Guests']].describe()
```

## d) Converted Hotel, Distribution Channel, Customer Type, Deposit Type, Country, Status → category

### Why: These are qualitative variables with repeated values. Using category:

#### By converting these columns to category, i'm explicitly telling pandas (and other people):

#### "These columns represent categories — not free-form text or numerical data."
##### This helps:

##### Prevent errors in analysis.

```python
categorical_cols = ['Hotel', 'Distribution Channel', 'Customer Type', 'Deposit Type', 'Country', 'Status']
SHG[categorical_cols] = SHG[categorical_cols].astype('category')
```

### 5) "Validating Business Logic"
#### checking if the data makes sense based on real-world rules or expectations.

#### a) If a booking is labeled "Canceled", it should have the binary cancellation flag set to 1.

###### Checked for mismatch

```python
SHG[SHG['Status'] == 'Canceled'][['Status', 'Cancelled (0/1)']].value_counts()
```

#### b) If Revenue = 0, is Revenue Loss > 0?
##### If the hotel made no revenue, it could mean:
###### The guest canceled last minute → so you lost money (Revenue Loss > 0).
###### Or the booking was valid but the room was free (Revenue Loss = 0).

```python
SHG[SHG['Revenue'] == 0][['Revenue Loss', 'Status']].value_counts()
```

#### c) Are there rows where Guests == 0?
##### You can't book a room with zero guests (usually).

```python
SHG[SHG['Guests'] == 0]
```

```python
SHG[SHG['Guests'] == 0]['Status'].value_counts()
```

```python
SHG[SHG['Guests'] == 0]['Cancelled (0/1)'].value_counts()
```

```python
SHG[SHG['Guests'] == 0][['Revenue', 'Revenue Loss']].describe()
```

# found out there are illogical reasonings 

* checked out bookings with zero guests
* some have non-zero revenue - which should not happen

## i dropped it completely because 
* they don't follow business logic
* they are few (180 out of 119,390) that's 0.15%
* they could mislead my analysis

```python
SHG = SHG[SHG['Guests'] != 0]
```
